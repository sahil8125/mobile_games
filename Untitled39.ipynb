```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Clean and convert Player count to numeric
df['Player count'] = df['Player count'].str.replace(',', '')
df['Player count (millions)'] = df['Player count'].str.extract(r'(\d+)').astype(float)
```

```python
# Convert to NumPy array
player_counts = df['Player count (millions)'].to_numpy()

# Get indices of top 10 games
top_indices = np.argsort(-player_counts)[:10]
top_games = df.iloc[top_indices]
```

```python
# Plot top 10 games
plt.figure(figsize=(10, 6))
plt.barh(top_games['Game Title'], top_games['Player count (millions)'], color='red')
plt.xlabel('Player Count (millions)')
plt.title('Top 10 Mobile Games by Player Count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
